```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
#import cleaned dataframe
clean_doc= pd.read_csv('drive/My Drive/Colab Notebooks/NBA Finals Sentiment Analysis/clean_doc.csv')
clean_doc
```

```python
from textblob import TextBlob

pol = lambda x: TextBlob(x).sentiment.polarity
sub = lambda x: TextBlob(x).sentiment.subjectivity

clean_doc['pol'] = clean_doc["0"].apply(pol)
clean_doc['subj'] = clean_doc["0"].apply(sub)

#data.to_csv('tweet_polarity.csv')
#clean_doc.reset_index(inplace=True)
#clean_doc.drop(clean_doc[1], axis=1)
```

```python
# Draw graphics
clean_doc = clean_doc.sort_values('pol')
plt.barh(clean_doc['id'], clean_doc['pol'])
plt.title("Polarity of July 5 Interview: Positive or Negaitve Language (scale of -1 to +1)")
plt.show()
```

```python
clean_doc = clean_doc.sort_values('subj')
plt.barh(clean_doc['id'], clean_doc['subj'])
plt.title("Subjectivity of July 5 Interview: The relative fact or opinion bias in the language used\
 (scale of 0=fact to +1=opinion)")
plt.show()
```
